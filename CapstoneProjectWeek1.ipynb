```python
results = requests.get(url).json()
results
```

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
venues = results['response']['groups'][0]['items']
    
Gachibowli_nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
Gachibowli_nearby_venues =Gachibowli_nearby_venues.loc[:, filtered_columns]

# filter the category for each row
Gachibowli_nearby_venues['venue.categories'] = Gachibowli_nearby_venues.apply(get_category_type, axis=1)

# clean columns
Gachibowli_nearby_venues.columns = [col.split(".")[-1] for col in Gachibowli_nearby_venues.columns]

Gachibowli_nearby_venues.head(10)
```

<h3>Map of Gachibowli,Hyderabad for reference to Manhattan</h3>

```python
# create map of Singapore place  using latitude and longitude values
map_hy = folium.Map(location=[latitude, longitude], zoom_start=20)

# add markers to map
for lat, lng, label in zip(Gachibowli_nearby_venues['lat'], Gachibowli_nearby_venues['lng'], Gachibowli_nearby_venues['name']):
    label = folium.Popup(label, parse_html=True)
    folium.RegularPolygonMarker(
        [lat, lng],
        number_of_sides=4,
        radius=10,
        popup=label,
        color='blue',
        fill_color='#0f0f0f',
        fill_opacity=0.7,
    ).add_to(map_hy)  
    
map_hy
```
